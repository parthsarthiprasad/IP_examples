```python
#Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img):                          # To increase picture size on Jupyter Lab
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img)

def load_img(x):                             # Loads image stored at address'x' in suitable format
    z=cv2.imread(x)
    if(type(z)!=None):                                                             #Checks if image exists at address given
        img=cv2.cvtColor(cv2.imread(x).astype(np.float32)/255,cv2.COLOR_BGR2RGB)
        return img
    print("Loading failed. Image was not found at the given address")      # Returns black image for wrong entry
    return np.zeros(shape=(200,200))
    
x=input("Enter address for the image")
```

```python
img=load_img(x)
show_img(img)
```

```python
# Median Blurring
median_blur_img = cv2.medianBlur(img,5) 
# 5 is the kernel size(ksize), should be odd and mainpulated according to image parameters(i.e. size) to gain optimal results
show_img(median_blur_img)
```

```python
# Gaussian Blurring
gauss_blur_img = cv2.GaussianBlur(img,(19,19),25) # (19,19) is kernel shape and 25 is standard deviation set along X-axis 
show_img(gauss_blur_img)
```

```python
# Bilateral Filter
bilateral_blur = cv2.bilateralFilter(img,29,50,50)
show_img(bilateral_blur)
```

```python
# Sharpening
sub = cv2.GaussianBlur(img,(0,0),3)
sharp_img=cv2.addWeighted(img,1.5,sub,-0.5,gamma=0)
show_img(sharp_img)
```
