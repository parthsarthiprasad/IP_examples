```python
# Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_TRIPLEX
    cv2.putText(blank_img,text='ARES',org=(50,300), fontFace=font,fontScale= 6,color=(255,255,255),thickness=20,lineType=cv2.LINE_AA)
    return blank_img

def show_img(img):
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img=load_img()
show_img(img)
```

```python
# Erosion
kernel = np.ones((5,5),np.uint8) #Creating kernel
ero_img = cv2.erode(img,kernel,iterations = 1)    # Apply erosion using single iteration 
show_img(ero_img)
```

```python
ero_img_1 = cv2.erode(img,kernel,iterations = 4)    # Apply erosion using 4 iteration
show_img(ero_img_1)
```

```python
#Dilation
dil_img=cv2.dilate(ero_img_1,kernel,iterations=4) # Dilating the eroded image
show_img(dil_img)
```

```python
# Opening - Often used to remove background noise
noise=np.random.randint(low=0,high=2,size=(600,600))*255 #creates random noise of value 255(white) or 0(black) of same size as image(600,600)
noise_img=img+noise
show_img(noise_img)
```

```python
open_img=cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel) # Applying Opening on noise image
show_img(open_img)
```

```python
# Closing - Often used to remove foreground image
noise_img=img-noise
noise_img[noise_img==-255]=0  #Creating Foregroung Noise
show_img(noise_img)
```

```python
close_img = cv2.morphologyEx(noise_img, cv2.MORPH_CLOSE, kernel) #Returns Cleared image
show_img(close_img)
```
