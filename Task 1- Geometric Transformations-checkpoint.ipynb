```python
#Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img,c=None):                          # To increase picture size on Jupyter Lab, c=cmap
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=c)

def load_img(x):                             # Loads image stored at address'x' in suitable format
    z=cv2.imread(x)
    if(type(z)!=None):                                                             #Checks if image exists at address given
        img=cv2.cvtColor(cv2.imread(x),cv2.COLOR_BGR2RGB)            # Transforming input image from BGR to RGB colour 
        return img
    print("Loading failed. Image was not found at the given address")      # Returns black image for wrong entry
    return np.zeros(shape=(200,200))
    
x=input("Enter address for the image")
```

```python
img=load_img(x)
show_img(img)
```

```python
# Scaling
scaling_factor_x=0.5 
scaling_factor_y=2
scaled_img=cv2.resize(img,None,fx=scaling_factor_x,fy=scaling_factor_y) # OR scaled_img=cv2.resize(img,dsize=(img.shape[0]*scaling_factor_y,img.shape[1]*scaling_factor_x))
show_img(scaled_img)
```

```python
# Translation
M = np.float32([[1,0,30],[0,1,175]]) # 30 pixels left,175 pixels down
dst = cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))
show_img(dst)
```

```python
# Rotation
# Upside Down
img_180=cv2.flip(cv2.flip(img,0),1)
show_img(img_180)
```

```python
# LEFT
M = cv2.getRotationMatrix2D((img.shape[0]/2,img.shape[1]/2),90,1)
image_left = cv2.warpAffine(img,M,(img.shape[0],img.shape[1]))
show_img(image_left)
```

```python
# Right
M = cv2.getRotationMatrix2D((img.shape[0]/2,img.shape[1]/2),-90,1)
image_left = cv2.warpAffine(img,M,(img.shape[0],img.shape[1]))
show_img(image_left)
```

```python
# Perspective Transformation
pts1 = np.float32([[30,20],[380,25],[25,450],[380,455]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
show_img(dst)
```
