```python
#Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img,c=None):                          # To increase picture size on Jupyter Lab, c=cmap
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=c)

def load_img(x):                             # Loads image stored at address'x' in suitable format
    z=cv2.imread(x)
    if(type(z)!=None):                                                             #Checks if image exists at address given
        img=cv2.imread(x,0)
        return img
    print("Loading failed. Image was not found at the given address")      # Returns black image for wrong entry
    return np.zeros(shape=(200,200))
    
x=input("Enter address for the image")
```

```python
img=load_img(x)
show_img(img,'gray')
```

```python
# Detecting Face
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
detected_img=img.copy()
faces=face_cascade.detectMultiScale(img,scaleFactor=1.1,minNeighbors=5) #Parameters must be adjusted for optimal effect.
for (x,y,w,h) in faces:                          
    cv2.rectangle(detected_img, (x,y), (x+w,y+h), (255,255,255), 10) # Creates a rectangle over the face
show_img(detected_img,'gray')
```

```python
# I seriously do not blame the program. Kisi ko bhi iski shakal pasand nahi aati -\('')/-

# Video
cap = cv2.VideoCapture(0) # Access the default camera

while True: 
    
    ret, frame = cap.read(0) 
    faces=face_cascade.detectMultiScale(frame,scaleFactor=1.1,minNeighbors=5) #Parameters must be adjusted for optimal effect.
    for (x,y,w,h) in faces:                          
        cv2.rectangle(frame, (x,y), (x+w,y+h), (255,255,255), 10) 
    cv2.imshow('Video Face Detection', frame) 
 
    c = cv2.waitKey(1) 
    if c == 27: # Exit on using Esc key
        break 
        
cap.release() 
cv2.destroyAllWindows()
```
