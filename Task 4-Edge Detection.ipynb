```python
# Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img,c=None):                          # To increase picture size on Jupyter Lab, c=cmap
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=c)

def load_img(x):                             # Loads image stored at address'x' in suitable format
    z=cv2.imread(x)
    if(type(z)!=None):                                                             #Checks if image exists at address given
        img=cv2.cvtColor(cv2.imread(x),cv2.COLOR_BGR2RGB)            # Transforming input image from BGR to RGB colour 
        return img
    print("Loading failed. Image was not found at the given address")      # Returns black image for wrong location entry
    return np.zeros(shape=(200,200))
    
x=input("Enter address for the image")
```

```python
img=load_img(x)
show_img(img)
```

```python
# Edge detection without Blurring - Often Leads to Noise
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
show_img(edges)
```

```python
# Blurring to avoid minor edges   Should be used for high def images
blur_img = cv2.blur(img,(5,5))
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
show_img(edges)
```

```python
# Optimizing thresholds For High Def Images Not suitable for current low def pic  
median_val = np.median(img) 
low=max((0,0.7*median_val))
high=min((255,1.3*median_val))
edges=cv2.Canny(img,threshold1=low,threshold2=high)
show_img(edges)
```
