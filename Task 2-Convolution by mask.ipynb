```python
# Importing Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img,c=None):                          # To increase picture size on Jupyter Lab, c=cmap
    fig = plt.figure(figsize=(15,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=c)

def load_img(x):                             # Loads image stored at address'x' in suitable format
    z=cv2.imread(x)
    if(type(z)!=None):                                                             #Checks if image exists at address given
        img=cv2.cvtColor(cv2.imread(x),cv2.COLOR_BGR2RGB)            # Transforming input image from BGR to RGB colour 
        return img
    print("Loading failed. Image was not found at the given address")      # Returns black image for wrong entry
    return np.zeros(shape=(200,200))
    
x=input("Enter address for the image")
```

```python
img=load_img(x)
show_img(img)
```

```python
mask=[[0,-1,0],[-1,5,-1],[0,-1,0]]  # Mask- Used for basic edge detection
new_img=np.zeros(shape=(img.shape))
for x in range(1,img.shape[1]-1):
    for y in range(1,img.shape[0]-1):
        new_img[y][x]=img[y-1][x-1]*mask[0][0]+img[y-1][x]*mask[0][1]+img[y-1][x+1]*mask[0][2]+img[y][x-1]*mask[1][0]+img[y][x]*mask[1][1]+img[y][x+1]*mask[1][2]+img[y+1][x-1]*mask[2][0]+img[y+1][x]*mask[2][1]+img[y+1][x+1]*mask[2][2]
show_img(new_img)
```
